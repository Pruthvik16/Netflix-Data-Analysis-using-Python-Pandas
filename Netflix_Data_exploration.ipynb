```python
#Q. 10) What is the maximum duration of a Movie on Netflix 

#know the data type
df.info()
```

```python
#see the unique values
df['Duration'].unique()
```

```python
# seperate the min to different columns for all the movies
all_the_movies = df['Type'].str.contains('Movie')
movies = df.loc[all_the_movies,:]
movies
```

```python
# seperate the duration and mintues in two different columns
movies[['duration','minutes']] = movies['Duration'].str.split(" ", expand = True)
movies
```

```python
#change the datatype to int
movies['duration'] = movies['duration'].astype(int)
```

```python
#find the max duration of the movie
movies['duration'].max()
```

```python
#find the min duration of the movie
movies['duration'].min()
```
